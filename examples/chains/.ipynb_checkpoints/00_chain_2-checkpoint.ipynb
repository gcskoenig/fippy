# Deriving analytic RFI value

### Multiple Linear regression

Following the http://dept.stat.lsa.umich.edu/~kshedden/Courses/Regression_Notes/prediction.pdf

Mean squared prediction error (MSPE) is defined as:

$$\text{MSPE} = \frac{1}{n^*} \mathbb{E} \left\lVert y^* - f_{\hat{\beta}}(X^∗)  \right\rVert^2 = \frac{1}{n^*} \sum_{i=1}^{n^*} \mathbb{E} \left(y^*_i - f_{\hat{\beta}}(x^∗_i)\right)^2,$$

where $y^* = X^∗\beta + \varepsilon$ is testing data, $\varepsilon \sim N(0, \sigma^2 I)$. $X^∗ \in Mat(n^* \times p)$ could be either random or fixed. From regression notes:

$$\text{MSPE} = \frac{1}{n^*} \text{tr}(X^{*T}X^* \Sigma_{\hat{\beta}}) + \sigma^2, \quad \text{where $\Sigma_{\hat{\beta}} = \sigma^2 (X^T X)^{-1}$}$$

It can be shown, that when keeping $\frac{X^{*T}X^*}{n^*} = \frac{X^T X}{n} = \hat{Cov}(x)$,   $\text{MSPE} \xrightarrow[n \to \infty]{} \sigma^2$.

To calculate a theoretic value of $\text{RFI}_j$, we need to find a risk with a resampled replacement variable $j$. Let $\tilde{X}^{*j}$ be a test data with conditionally resampled variable $j$. Then Resampled Mean squared prediction error ($\tilde{\text{MSPE}_j}$) is:

$$\tilde{\text{MSPE}_j} = 
  \frac{1}{n^*} \mathbb{E} \left\lVert y^* − \tilde{X}^{∗j} \hat{\beta}  \right\rVert^2 = 
  \frac{1}{n^*} \mathbb{E} \left\lVert X^* \beta − \tilde{X}^{∗j} \hat{\beta}  \right\rVert^2 + \frac{1}{n^*} \mathbb{E} \left\lVert \varepsilon \right\rVert^2 = 
  \frac{1}{n^*} \sum_{i=1}^{n^*}  \mathbb{E} (x_i^{*T} \beta - \tilde{x}_i^{*T} \hat{\beta})^2 + \sigma^2$$

$$\mathbb{E} (x_i^{*T} \beta - \tilde{x}_i^{*T} \hat{\beta})^2 = 
  \mathbb{E} (x_{i,j}^{*} \beta_j + x_{i,-j}^{*T} \beta_{-j} - \tilde{x}_{i,j}^{*} \hat{\beta}_j  - x_{i,-j}^{*T} \hat{\beta}_{-j})^2 = 
  \mathbb{E} \left[ (x_{i,j}^{*} \beta_j - \tilde{x}_{i,j}^{*} \hat{\beta}_j)^2 + (x_{i,-j}^{*T} (\beta_{-j} - \hat{\beta}_{-j}))^2  - 2 (x_{i,j}^{*} \beta_j - \tilde{x}_{i,j}^{*} \hat{\beta}_j)(x_{i,-j}^{*T} (\beta_{-j} - \hat{\beta}_{-j}))\right] = \\
  = *_1 + *_2 + *_3$$
  
$$*_1 = \mathbb{E} (x_{i,j}^{*} \beta_j - \tilde{x}_{i,j}^{*} \hat{\beta}_j)^2 = 
  \mathbb{E} (x_{i,j}^{*} \beta_j)^2 + \mathbb{E} (\tilde{x}_{i,j}^{*} \hat{\beta}_j)^2 - 2 \mathbb{E} (x_{i,j}^{*} \beta_j) (\tilde{x}_{i,j}^{*}\hat{\beta}_j) = 
  (x_{i,j}^{*} \beta_j)^2 + \tilde{x}_{i,j}^{*2} \left(Var(\hat{\beta}_j) + \beta_j^2 \right) - 2 x_{i,j}^{*} \tilde{x}_{i,j}^{*}\beta_j ^2 = \\
  = \beta_j^2 (x_{i,j}^{*2} +  \tilde{x}_{i,j}^{*2} - 2 x_{i,j}^{*} \tilde{x}_{i,j}) + \tilde{x}_{i,j}^{*2} Var(\hat{\beta}_j) = \beta_j^2 (x_{i,j}^{*} - \tilde{x}_{i,j}^{*})^2 + \tilde{x}_{i,j}^{*2} Var(\hat{\beta}_j)$$

$$*_2 = \mathbb{E} (x_{i,-j}^{*T} (\beta_{-j} - \hat{\beta}_{-j}))^2 = 
  [\text{ ... following tutorial ...}] $$
  
$$*_3 = - 2 \mathbb{E} (x_{i,j}^{*} \beta_j - \tilde{x}_{i,j}^{*} \hat{\beta}_j)(x_{i,-j}^{*T} (\beta_{-j} - \hat{\beta}_{-j})) = [\text{some calculations ...}] = -2 \tilde{x}_{i,j}^{*} (x_{i,-j}^{*T} \Sigma_{\hat{\beta}_{j,-j}}) $$

Putting everything together and transforming to matrices:

$$ \tilde{\text{MSPE}_j} = \frac{\beta_j^2}{n^*} \left\lVert X^*_{:, j} - \tilde{X}^*_{:, j} \right\rVert^2 + \frac{Var(\hat{\beta}_j)}{n^*} \left\lVert \tilde{X}^*_{:, j} \right\rVert^2 - \frac{2}{n^*} \tilde{X}^{*T}_{:, j} X^*_{:,-j} \Sigma_{\hat{\beta}_{j, -j}} + \frac{1}{n^*}\text{tr}(X^{*T}_{:,-j}X^*_{:,-j} \Sigma_{\hat{\beta}_{-j,-j}}) + \sigma^2$$

By keeping $\frac{X^{*T}X^*}{n^*} = \frac{X^T X}{n} = \hat{Cov}(x)$ and increasing $n$:  $\frac{2}{n^*} \tilde{X}^{*T}_{:, j} X^*_{:,-j} \Sigma_{\hat{\beta}_{j, -j}} + \frac{1}{n^*}\text{tr}(X^{*T}_{:,-j}X^*_{:,-j} \Sigma_{\hat{\beta}_{-j,-j}})$ will tend to $0$.

Also, we have a fomula for $Var(\hat{\beta}_j) = \frac{\sigma^2}{(1 - R^2)(\sum_{i=1}^n (x_{ij} - \overline{x_j})}$. It also tends to $0$, if we increase $n$.

Now, let's calculate the risk for large enough $n$:

$$\mathbb{E}_{\tilde{x}_{i,j} \sim P(X_j / G), x_{i} \sim P(X_1, ... X_P) } \tilde{\text{MSPE}_j} 
  \approx \beta_j^2 \mathbb{E}(\tilde{X}_j - X_j)^2 + \sigma^2 = 
  \beta_j^2 \left(\mathbb{E}\tilde{X}_j^2 - 2 \mathbb{E} X_j \tilde{X}_j + \mathbb{E} X_j^2 \right) + \sigma^2 = 
  \beta_j^2 \left( \text{Var} (\tilde{X}_j) +  (\mathbb{E} \tilde{X}_j)^2  + \text{Var} ({X}_j) +  (\mathbb{E} {X}_j)^2 - 2 \text{Cov}(\tilde{X}_j, {X}_j) - 2 \mathbb{E}{X}_j \mathbb{E}\tilde{X}_j \right) + \sigma^2 = \\
  = 2 \beta_j^2 \left(\text{Var}(X_j) - \text{Cov}(\tilde{X}_j, {X}_j) \right) + \sigma^2 $$
  
Finally, RFI for large $n$:

$$\text{RFI}_j = \mathbb{E}_{\tilde{x}_{i,j} \sim P(X_j / G), x_{i} \sim P(X_1, ... X_P) } \left[ \tilde{\text{MSPE}_j} - \text{MSPE} \right] \approx 2 \beta_j^2 \left(\text{Var}(X_j) - \text{Cov}(\tilde{X}_j, {X}_j) \right)$$

### Chain 2 Example



$Y = X_3 + X_4 + \varepsilon_Y, \quad \varepsilon_Y \sim N(0, 0.25)$

$(X_1, X_2, X_3, X_4) \sim N \left(0, \begin{pmatrix} 1 & 1 & 1 & 1 \\ 1 & 2 & 2 & 1 \\ 1 & 2 & 2.3 & 1 \\ 1 & 1 & 1 & 2\end{pmatrix} \right)$

$ \tilde{X}_{j} / X_{G} = ε * \sqrt{𝛴_{jj} - 𝛴_{jG} 𝛴_{GG}^{-1} 𝛴_{Gj} }+ 𝛴_{jG} 𝛴_{GG}^{-1} X_G, \quad 𝜀 \sim N(0, 1), \quad 𝜀 \text{ - independent of everything}$

$\text{Cov}(\tilde{X}_j, {X}_j) = \text{Cov}(𝛴_{jG} 𝛴_{GG}^{-1} X_G, X_j) = 𝛴_{jG} 𝛴_{GG}^{-1} 𝛴_{Gj}$

$\text{RFI}_j \approx 2 \beta_j^2 \left(𝛴_{jj} - 𝛴_{jG} 𝛴_{GG}^{-1} 𝛴_{Gj} \right) $

```python
sigma_y = 0.5
Sigma_X = np.array(((1, 1, 1, 1), (1, 2, 2, 1), (1, 2,  2.09, 1), (1, 1,  1, 2)), dtype=float)
beta = np.array((0, 0, 1, 1))

def MSPE_difference(X_train, X_test, beta, sigma_y, j, perturbed_foi):
    n_star = len(X_test)
    
    Sigma_beta_hat = sigma_y**2 * np.linalg.inv(X_train.T @ X_train)
    MSPE = np.diag(X_test.T @ X_test @ Sigma_beta_hat).sum() / n_star + sigma_y**2
    
    X_test_minus_j = np.concatenate((X_test[:, :j], X_test[:, j+1:]), axis=1)
    Sigma_beta_hat_j_minus_j = np.concatenate((Sigma_beta_hat[j, :j], Sigma_beta_hat[j, j+1:]))
    Sigma_beta_hat_minus_j_minus_j = np.concatenate(
        (np.concatenate((Sigma_beta_hat[:j, :j], Sigma_beta_hat[:j, j+1:]), axis=1),
        np.concatenate((Sigma_beta_hat[j+1:, :j], Sigma_beta_hat[j+1:, j+1:]), axis=1)),
        axis=0
    )
    
    tilde_MSPE_j = beta[j]**2 * np.linalg.norm(X_test[:, j] - perturbed_foi)**2 / n_star + \
        Sigma_beta_hat[j, j] * np.linalg.norm(perturbed_foi)**2 / n_star - \
        2/n_star*perturbed_foi.T @ X_test_minus_j @ Sigma_beta_hat_j_minus_j + \
        np.diag(X_test_minus_j.T @ X_test_minus_j @ Sigma_beta_hat_minus_j_minus_j).sum() / n_star + sigma_y**2
    
#     print(beta[j]**2 * np.linalg.norm(X_test[:, j] - perturbed_foi)**2 / n_star, 
#          Sigma_beta_hat[j, j] * np.linalg.norm(perturbed_foi)**2 / n_star, 
#          2/n_star*perturbed_foi.T @ X_test_minus_j @ Sigma_beta_hat_j_minus_j,
#          np.diag(X_test_minus_j.T @ X_test_minus_j @ Sigma_beta_hat_minus_j_minus_j).sum() / n_star)
    
#     print(np.cov([X_test[:, j], perturbed_foi], rowvar=True))
    return tilde_MSPE_j - MSPE

def RFI_approx(Sigma_X, beta, j, G):
#     print(Sigma_X[j, j], 
#          Sigma_X[j, G] @ np.linalg.inv(Sigma_X[G, G].reshape(len(G), len(G))) @ Sigma_X[G, j])
    RFI = 2 * beta[j]**2 * (Sigma_X[j, j] - Sigma_X[j, G] @ np.linalg.inv(Sigma_X[np.ix_(G,G)].reshape(len(G), len(G))) @ Sigma_X[G, j])
    return RFI

def plot_RFI_GT_and_MC(X_train, X_test, beta, sigma_y, perturbed_foiss, ax, Sigma_X, fsoi, G):
    
    RFI_MC = []
    RFI_GT = []
    nr_runs = perturbed_foiss.shape[1]

    for j in fsoi:
        RFI_MC.append(np.mean([MSPE_difference(X_train, X_test, beta, sigma_y, j, perturbed_foiss[j, run, :]) 
                                    for run in range(nr_runs)]))
        RFI_GT.append(RFI_approx(Sigma_X, beta, j, G))
    
    ax.scatter(x = RFI_MC, y = [ax.patches[i].get_xy()[1] + ax.patches[i].get_height()/2 for i in fsoi],
           zorder=2, label = 'MC RFI estimate')
    ax.scatter(x = RFI_GT, y = [ax.patches[i].get_xy()[1] + ax.patches[i].get_height()/2 for i in fsoi],
               zorder=2, label = 'Approx GT RFI', marker='^', s=100, color='black')
```

```python
# LOADING DATA AND FITTING MODEL
name = 'chain2'

N= 10**5
dataset = np.loadtxt('data/{}.csv'.format(name), dtype=np.float32)

splitpoint = math.floor(N*0.9)
ix_train = np.arange(0, splitpoint, 1)
ix_test = np.arange(splitpoint, N, 1)

X_train, y_train = dataset[ix_train, :-1], dataset[ix_train,-1]
X_test, y_test = dataset[ix_test, :-1], dataset[ix_test,-1]

model = LinearRegression()
model.fit(X_train, y_train)
names = ['x1', 'x2', 'x3', 'x4', 'y']
print(names[0:-1])
print(model.coef_)
#0.00297259 -0.00854544  1.0052813   1.0001918

y_pred = model.predict(X_test)
risk = mean_squared_error(y_test, y_pred)
```

### G = [0] 

```python
G = np.array([0])
fsoi = np.array([0, 1, 2, 3], dtype=np.int16)
names = np.array(names)
```

```python
gauss_sampler = gaussian.GaussianSampler(X_train)
gauss_sampler.train(fsoi, G)

rfi_gauss_explainer = explainer.Explainer(model.predict, fsoi, X_train, sampler=gauss_sampler, 
                                          loss=mean_squared_error, fs_names=names)
gauss_explanation, perturbed_foiss = rfi_gauss_explainer.rfi(X_test, y_test, G, nr_runs=100)

    
ax = gauss_explanation.barplot()
plot_RFI_GT_and_MC(X_train, X_test, beta, sigma_y, perturbed_foiss, ax, Sigma_X, fsoi, G)
plt.legend()
plt.title(f'Gaussian sampler. G = {G}, N = {N}')
```

```python
nflow_sampler = cnflow.CNFSampler(X_train)
nflow_sampler.train(fsoi, G)

rfi_gauss_explainer = explainer.Explainer(model.predict, fsoi, X_train, sampler=gauss_sampler, 
                                          loss=mean_squared_error, fs_names=names)
rfi_nflow_explainer = explainer.Explainer(model.predict, fsoi, X_train, sampler=nflow_sampler, 
                                          loss=mean_squared_error, fs_names=names)

nflow_explanation, perturbed_foiss = rfi_nflow_explainer.rfi(X_test, y_test, G)
ax = nflow_explanation.barplot()
plot_RFI_GT_and_MC(X_train, X_test, beta, sigma_y, perturbed_foiss, ax, Sigma_X, fsoi, G)
plt.legend()
plt.title(f'CNF sampler. G = {G}')
```
